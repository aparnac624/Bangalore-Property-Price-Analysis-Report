# Detecting the Outliers

```python
# Q2. Detect the outliers using following methods and remove it using methods like trimming / capping/ imputation using mean or median 
# a) Mean and Standard deviation
# calculating the mean of column price_per_sqft
mean = df_cleaned_data['price_per_sqft'].mean()
# calculating the standard deviation of column price_per_sqft
std = df_cleaned_data['price_per_sqft'].std()

# setting the upper and lower limits
upper_limit = mean + 3*std
lower_limit = mean - 3*std

# trimming the outliers and adding them to a new dataframe 
df_trimmed_std = df_cleaned_data[(df_cleaned_data['price_per_sqft'] >= lower_limit) & (df_cleaned_data['price_per_sqft'] <= upper_limit)]
df_trimmed_std
```

```python
# b)Percentile method
# calculating the fifth percentile
lower_percentile = df_cleaned_data['price_per_sqft'].quantile(0.05)
# calculating the 95th percentile
upper_percentile = df_cleaned_data['price_per_sqft'].quantile(0.95)

# trimming the outliers and add them to a new dataframe 
df_trimmed_percentile = df_cleaned_data[(df_cleaned_data['price_per_sqft'] >= lower_percentile) & (df_cleaned_data['price_per_sqft'] <= upper_percentile)]
df_trimmed_percentile
```

```python
# c) IQR(Inter quartile range method) 
# calculating the first quantile
Q1 = df_cleaned_data['price_per_sqft'].quantile(0.25)
# calculating the third quantile
Q3 = df_cleaned_data['price_per_sqft'].quantile(0.75)
IQR = Q3 - Q1

# setting the upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# trimming the outliers and adding it to a new dataframe
df_trimmed_iqr = df_cleaned_data[(df_cleaned_data['price_per_sqft'] >= lower_bound) & (df_cleaned_data['price_per_sqft'] <= upper_bound)]
df_trimmed_iqr
```

```python
# d) Z Score method
# importing the zscore function
from scipy.stats import zscore

# Creating a copy to avoid modifying the original
df_z = df_cleaned_data.copy()

# calculating the z-score of each value in price_per_sqft column
df_z['z_score'] = zscore(df_z['price_per_sqft'])

# filtering values with z-score bteween -3 and +3
# values outside this range are considered outliers
df_trimmed_z = df_z[(df_z['z_score'] >= -3) & (df_z['z_score'] <= 3)]
df_trimmed_z
```

# Box Plot - Determines which method works best in removing outliers for the data

```python
# Q3. Creating a box plot and using this to determine which method seems to work best to remove outliers for this data 
import matplotlib.pyplot as plt
import seaborn as sns

# setting the size of plot to 15 inches wide and 6 inches tall
plt.figure(figsize=(15, 6))

# creating the box plot 
sns.boxplot(data=[df_cleaned_data['price_per_sqft'], 
                  df_trimmed_std['price_per_sqft'], 
                  df_trimmed_percentile['price_per_sqft'], 
                  df_trimmed_iqr['price_per_sqft'], 
                  df_trimmed_z['price_per_sqft']],
                  orient="h")

# adding the labels and titles
plt.xlabel("Price per Sqft")
plt.title("Boxplot for Original and Cleaned Data (Methods: Original, Std, Percentile, IQR, Z-score)")

# shows the plot
plt.show()
```

# Histogram Plots

```python
# Q4. Drawing histplot to check the normality of the column(price per sqft column).
# Checking the skewness and kurtosis before and after the transformation.
from scipy.stats import skew, kurtosis
from scipy.stats import boxcox
import numpy as np

# Histogram before transformation
sns.histplot(df_trimmed_iqr['price_per_sqft'],bins = 30, kde=True)
plt.title('Before Transformation')
plt.xlabel("Price per Sqft")
plt.ylabel("Count")
plt.show()
```

```python
# Skewness & Kurtosis before transformation
print("Skewness (Before):", skew(df_trimmed_iqr['price_per_sqft']))
print("Kurtosis (Before):", kurtosis(df_trimmed_iqr['price_per_sqft']))
```

```python
# Applying log1p transformation 
df_transformed = df_trimmed_iqr.copy()
df_transformed['log_price_per_sqft'] = np.log1p(df_transformed['price_per_sqft'])
```

```python
# Histogram after transformation
sns.histplot(df_transformed['log_price_per_sqft'], bins=30, kde=True)
plt.title('After Log Transformation')
plt.xlabel("Log(Price per Sqft)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Skewness and Kurtosis after transformation
print("Skewness (After):", skew(df_transformed['log_price_per_sqft']))
print("Kurtosis (After):", kurtosis(df_transformed['log_price_per_sqft']))
```

# Correlation between all numerical columns - Plotting Heatmap

```python
# Q5. Checking the correlation between all the numerical columns and plot heatmap. 
# Selecting numerical columns from the cleaned and transformed dataset
num_df = df_transformed.select_dtypes(include=['int64', 'float64'])

# calculating the Correlation matrix
corr = num_df.corr()

# visualizing using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidth=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
